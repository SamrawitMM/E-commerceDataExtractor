```python
import pandas as pd
import re
```

```python
df = pd.read_csv('/data/telegram_data.csv')
```

```python
df.head()
```

```python
# Remove Rows with Missing Messages
df = df[df['Message'].notnull()].reset_index(drop=True)
print("After dropping empty messages:", df.shape)

# Preprocess Amharic Text Messages


def preprocess_amharic(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emojis and symbols (including any character not in Amharic range or space/punctuation)
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002702-\U000027B0"  # other symbols
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Keep only Amharic letters, basic punctuation, and spaces (optional: allow numbers too)
    text = re.sub(r'[^\u1200-\u137F\s፡።፣፤፥፦፧፨]', '', text)

    # Normalize whitespace
    text = re.sub(r'\s+', ' ', text)

    return text.strip()


# Apply preprocessing
df['Clean_Message'] = df['Message'].apply(preprocess_amharic)

# Select Structured Columns
structured_df = df[['Channel Title', 'Channel Username', 'ID', 'Date', 'Clean_Message', 'Media Path']]
structured_df.head()

# Save Preprocessed Data to CSV
structured_df.to_csv("preprocessed_telegram_data.csv", index=False)
print("✅ Preprocessed data saved to 'preprocessed_telegram_data.csv'")
```
